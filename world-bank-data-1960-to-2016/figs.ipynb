```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "fertility_rate.csv"

# Read Purchasing File and store into Pandas data frame
fertility_data = pd.read_csv(file_to_load)
fertility_data
```

```python
from matplotlib import style
from matplotlib import pyplot as plt
%matplotlib inline
style.use('ggplot')
```

```python
fertility_data=fertility_data.drop(['Country Code','Indicator Code','Indicator Name'], axis=1)
fertility_data.head()
```

```python
#Fig1

objects = (fertility_data['Country Name'])
y_pos = np.arange(len(objects))
width = .5

year_1960 = sorted(fertility_data['1960'])
year_2016 = sorted(fertility_data['2016'])


plt.figure(figsize=(10,10))

plt.barh(y_pos +width, year_1960, align='center',  label='1960')
plt.barh(y_pos +width, year_2016, align='center',  label='2016')



plt.yticks(y_pos, objects, fontsize='small')
plt.xlabel('Percentage')
plt.title('FIGURE 1. Fertility Worldwide in 1960 vs 2016.')

plt.legend()

plt.savefig('Fig1.jpg')

plt.show()
```
