```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load 
file_to_load = "fertility_rate.csv"

# Read Purchasing File and store into Pandas data frame
fertility_data = pd.read_csv(file_to_load)
fertility_data
```

```python
from matplotlib import style
from matplotlib import pyplot as plt
%matplotlib inline
style.use('ggplot')
```

```python
fertility_data=fertility_data.drop(['Country Code','Indicator Code','Indicator Name'], axis=1)
fertility_data.head()
```

```python
objects = (fertility_data['Country Name'])
y_pos = np.arange(len(objects))
width = .5

year_1960 = sorted(fertility_data['1990'])
year_2016 = sorted(fertility_data['2015'])


plt.figure(figsize=(10,10))

plt.barh(y_pos +width, year_1960, align='center',  label='1960')
plt.barh(y_pos +width, year_2016, align='center',  label='2016')



plt.yticks(y_pos, objects, fontsize='small')
plt.xlabel('Percentage')
plt.title('FIGURE 1. Fertility Worldwide in 1960 vs 2016.')

plt.legend()

plt.show()

plt.savefig('"FIGURE 1.png", dpi=600)
```

```python
# File to Load 
file_to_load = "data_ori.csv"

# Read Purchasing File and store into Pandas data frame
country_data = pd.read_csv(file_to_load)
country_data.head()
```

```python
country_data.rename(columns=
                    {'code':'Country Code'},
                    inplace = True
                )
country_data.head()
```

```python
fertility_country = pd.merge(left=fertility_data,right=country_data, how='left', left_on='Country Code', right_on='Country Code')
fertility_country.head()
```

```python
fertility_df = fertility_country[['Country Name','country','2016','Country Code','region','lat','long']]
fertility_df
```

```python
fertility_df.index = np.arange(1, len(fertility_df) + 1)
fertility_df.head()
```

```python
fertility_df.index.name = 'New ID'
fertility_df.head()
```

```python
fertility_df.loc[:,'2016'] *= 4
fertility_df
```

```python
fertility_df.rename(columns=
                    {'2016':'birth'},
                    inplace = True
                )
fertility_df.head()
```

```python
fertility_df.rename(columns=
                    {'Country Code':'code'},
                    inplace = True
                )
fertility_df.head()
```

```python
fertility_df.to_csv('data.csv',header=True)
```
